```python
E_bethe = -88.6335532522416
E_bethe_list.append(E_bethe)
E_bethe_list
```

```python
data_dir = os.path.join('L_200/')
nw_l = []
enw = []
enw_err = []

#max projection lenght to trust
pfix = 30 
#max p.l. to plot 
maxP = 100


for nw in nw_list:
    energy_list =[]
    err_list = []
    p_list = []
    count = 0
    with open(os.path.join(data_dir + 'L200_nw{}.dat'.format(nw)),'r') as f:
        for line in f: 
            count +=1 
            data = [ float(x.strip('\n')) for x in line.split('\t') if x.strip('')]
            p_list.append(data[0])
            energy_list.append(data[1])
            err_list.append(data[2])
            if count == pfix: 
                nw_l.append(1.0/nw)
                enw.append(data[1])
                enw_err.append(data[2])
    plt.figure(figsize=(8,5))
    plt.xlabel('Projection lenght p',fontsize = 12)
    plt.ylabel('Energy',fontsize = 12)
    plt.xlim(0,maxP)
    plt.xticks(range(0,maxP + 1,10))
    plt.errorbar(p_list,energy_list,yerr = err_list, color = 'black',label = 'nw = {}'.format(nw))
    plt.hlines(E_bethe,1,maxP,color = 'red', label = 'Bethe ansatz')
    plt.legend(fontsize = 12)
    plt.show()
e_nw_vs_L.append(enw)
#print(e_nw_vs_L)
```

```python
energy_list =[]
err_list = []
p_list = []
with open(os.path.join(data_dir + 'L200_1mil_bin300.dat'),'r') as f:
    for line in f: 
        data = [ float(x.strip('\n')) for x in line.split('\t') if x.strip('')]
        p_list.append(data[0])
        energy_list.append(data[1])
        err_list.append(data[2])
plt.figure(figsize=(8,5))
plt.xlabel('Projection lenght p',fontsize = 12)
plt.ylabel('Energy',fontsize = 12)
plt.xlim(0,maxP)
plt.xticks(range(0,maxP + 1,10))
plt.errorbar(p_list,energy_list,yerr = err_list, color = 'black',label = 'nw = {}'.format(nw))
plt.hlines(E_bethe,1,maxP,color = 'red', label = 'Bethe ansatz')
plt.legend(fontsize = 12)
plt.show()
```

Let's fix the projection time $p = 30$ ( notice that $ p \times n_{bra} = 150$ ) and plot the energy as a function of $ 1/ N_w$. 

```python
plt.figure(figsize=(8,5))
plt.xlabel(r'$ 1 / N_w$',fontsize = 12)
plt.ylabel('Energy',fontsize = 12)
plt.errorbar(nw_l,enw,yerr = enw_err,fmt = '*', color = 'black',label = r'Energy ($p = 30$)')
plt.plot(nw_l, enw, linewidth = 0.3,color ='black')
plt.hlines(E_bethe,0.0,0.01,color = 'red', label = 'Bethe ansatz')
plt.legend(fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.xlabel(r'$ 1 / N_w$',fontsize = 12)
plt.xscale('log')
plt.ylabel('Energy',fontsize = 12)
plt.errorbar(nw_l,enw,yerr = enw_err,fmt = '*', color = 'black',label = r'Energy ($p = 30$)')
plt.plot(nw_l, enw, linewidth = 0.3,color ='black')
plt.hlines(E_bethe,0.0,0.01,color = 'red', label = 'Bethe ansatz')
plt.legend(fontsize = 12)
plt.show()
```
